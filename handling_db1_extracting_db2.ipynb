## 필요한 열(Columns) 추출
- index , '출품작', '감독'

```python
# ======================================(4-1): 의미있는 col 만 추출
db2 = db2_raw[['출품작', '감독']]
db2
```

## 감독 column NaN 값이면 제거
- NaN 처리 방법 고민 전에 우선 BE 에 넘겨야 해서 관련 행 제거하고 진행

```python
# ======================================(4-2): NaN(결측값) 행 제거
db2_nan = db2.dropna(axis=0)
db2_nan
```

```python
# ======================================(): column 명 영문 변경
db2_nan.columns = ['title', 'director']
db2_nan
```

```python
# ======================================(5-1): csv 로 저장
db2.to_csv("db2_title_2021.csv", index=True, header=True)
```

```python
# ======================================(5-2): csv 로 저장
db2_nan.to_csv("db2_title_NaN_2021.csv", index=True, header=True)
```
